```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="PetalLengthCm", y="PetalWidthCm", hue="Species", palette="Set2")
plt.title("Scatter Plot: Petal Length vs Petal Width")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df["SepalLengthCm"], bins=20, kde=True, color="skyblue")
plt.title("Histogram: Sepal Length Distribution")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="Species", y="PetalWidthCm", palette="Set3")
plt.title("Box Plot: Petal Width by Species")
plt.xlabel("Species")
plt.ylabel("Petal Width (cm)")
plt.show()
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Iris.csv")

# Inspect the dataset
print("Shape of dataset:", df.shape)
print("Column names:", df.columns)
print("First 5 rows:")
print(df.head())
```

```python
# Summary statistics
print("\nDescriptive statistics:")
print(df.describe())

# Check for null values
print("\nMissing values:")
print(df.isnull().sum())

# Check class distribution
print("\nSpecies value counts:")
print(df["Species"].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")  # Set visual style
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="PetalLengthCm", y="PetalWidthCm", hue="Species", palette="Set2")
plt.title("Scatter Plot: Petal Length vs Petal Width")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.show()


plt.figure(figsize=(8, 6))
sns.histplot(df["SepalLengthCm"], bins=20, kde=True, color="skyblue")
plt.title("Histogram: Sepal Length Distribution")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()


plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="Species", y="PetalWidthCm", palette="Set3")
plt.title("Box Plot: Petal Width by Species")
plt.xlabel("Species")
plt.ylabel("Petal Width (cm)")
plt.show()
```
